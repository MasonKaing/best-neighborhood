# The total arrest count  
we calculated the total arrest count as a metric of if a neibourhood is good, since a low crime rate means a safer life.   
The neibourhood with lowest arrest counts are as below.

```python
import pandas as pd
df = pd.read_excel('arrests_2024_to_sept_2025.xlsx')
result_df = df['Neighborhood'].value_counts().reset_index()
result_df.columns = ['Neighborhood', 'Count']
print(result_df.tail(10))
#output_filename = 'arrests_summary.xlsx'
#result_df.to_excel(output_filename, index=False)
```
